#Bert_emb

```python
import os

os.environ['CUDA_LAUNCH_BLOCKING'] = "0" 
os.environ['CUDA_VISIBLE_DEVICES'] = "0" 

import timm
import torch.nn as nn
import torch
from PIL import Image
import requests
from torchvision import transforms
from torchvision.transforms import functional as FF
import math
from numpy import dot
from numpy.linalg import norm

from IPython.display import clear_output
import io
import base64
from IPython.display import HTML

def cosine_similarity(list_1, list_2):
  cos_sim = dot(list_1, list_2) / (norm(list_1) * norm(list_2))
  return cos_sim

  
EMB_SIZE = 1024

class Encoder(nn.Module):#beit_base_patch16_224_in22k,  beitv2_large_patch16_224_in22k
    
    def __init__(self):
        super().__init__()
        self.backbone_beit = timm.create_model('beit_large_patch16_224_in22k', pretrained=True, num_classes=0)
        if torch.cuda.is_available():
            self.backbone_beit.cuda()
        self.avgpool1d = nn.AdaptiveAvgPool1d(EMB_SIZE)
        
    def forward(self, x):
        x = FF.resize(x, size=[224, 224])
        x = x / 255.0
        x = x.type(torch.float32)
        outputs = self.backbone_beit(x)
        embedding = self.avgpool1d(outputs)
        
        return embedding
    
model_beit = Encoder()
model_beit.eval()
```

#FacePose

```python
import numpy as np
import cv2
from imutils import face_utils
import dlib
    
predictor = dlib.shape_predictor('/content/shape_predictor_68_face_landmarks.dat')
detector = dlib.get_frontal_face_detector()
image_size = 224

def facePose(img):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) 
    rects = detector(gray, 1)
    return rects

def goface(img, rects):
    # load image
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)    
    
    if True: #for rect in rects:
        shape = predictor(gray, rects)
        shape = face_utils.shape_to_np(shape)      
        shape = np.round(shape)
              
        # draw mask 
        msk = np.zeros(img.shape, dtype=np.uint8)
        cv2.fillPoly(msk, [cv2.convexHull(shape)], (1,1,1))
              
        # crop & resize
        umin = np.min(shape[:,0]) 
        umax = np.max(shape[:,0])
        vmin = np.min(shape[:,1]) 
        vmax = np.max(shape[:,1])
                  
        umean = np.mean((umin,umax))
        vmean = np.mean((vmin,vmax))
                          
        l = round( 1.2 * np.max((umax-umin,vmax-vmin)))
              
        if (l > np.min(img.shape[:2])):
            l = np.min(img.shape[:2])
                
        us = round(np.max((0,umean-float(l)/2)))
        ue = us + l
            
        vs = round(np.max((0,vmean-float(l)/2))) 
        ve = vs + l
                    
        if (ue>img.shape[1]):
            ue = img.shape[1]
            us = img.shape[1]-l
            
        if (ve>img.shape[0]):
            ve = img.shape[0]
            vs = img.shape[0]-l
                        
        us = int(us)
        ue = int(ue)  
              
        vs = int(vs)
        ve = int(ve)    
              
        img_cut = cv2.resize(img[vs:ve,us:ue],(256,256))
        msk = cv2.resize(msk[vs:ve,us:ue],(256,256),interpolation=cv2.INTER_NEAREST)
              

        mask = msk* 255

        # Load image, grayscale, Otsu's threshold
        image = img_cut
        mask = cv2.cvtColor(mask, cv2.COLOR_BGR2GRAY)
        #mask = cv2.dilate(mask, cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (50,50)))

        # Remove background using bitwise-and operation
        result = cv2.bitwise_and(image, image, mask=mask)
        #result[mask==0] = [255,255,255] # Turn background white
        result = cv2.resize(img_cut, (image_size, image_size), interpolation=cv2.INTER_AREA)
        return result
```

```python
def get_load2(img):
    img = Image.fromarray(img[..., ::-1]) 
    convert_to_tensor = transforms.Compose([transforms.PILToTensor()])
    input_tensor = convert_to_tensor(img)
    #input_tensor = input_tensor.unsqueeze(0)
    return input_tensor
```

```python
def get_embedding_bert(img):
    img = Image.fromarray(img[..., ::-1]) 
    convert_to_tensor = transforms.Compose([transforms.PILToTensor()])
    input_tensor = convert_to_tensor(img)
    input_batch = input_tensor.unsqueeze(0)
    with torch.no_grad():
        embedding = torch.flatten(model_beit(input_batch.cuda())[0]).cpu().data.numpy()
    return embedding
```

```python
def get_embedding2(imglist):
    with torch.no_grad():
        embedding = model_beit(torch.stack(imglist).cuda()).cpu().data.numpy()
    return embedding
```

```python
def classification(emb):
    pred = model_ml.predict_proba(emb)
    best_class = np.argmax(pred, axis=1)[0]
    acc  = pred[0][best_class]
    name = unique_names[best_class]
    print(f"Similar to: {name}, ACC: {acc:.2f}")
    # print(pred)
    return name, acc

#font = cv2.FONT_HERSHEY_DUPLEX
font = 5 #cv2.FONT_HERSHEY_COMPLEX

def test_one(path_img):
    img = cv2.imread(path_img)
    draw_img = img.copy()
    all_boxes = face_detect(img, one_only=False)
    norf = all_boxes.shape[0]
    for box in all_boxes:
        cut_fc = load_mtcnn2.Crop_Align(img, box)
        #embed_face = FACENET.Embedding([cut_fc], MTCNN.models)
        embed_face = FACENET.Embedding(cut_fc, norf, MTCNN.models)

        name, acc = classification(embed_face)
        
        x,y,x2,y2,_ = box.astype(int)
        
        
        
        cv2.rectangle(draw_img, (int(x), int(y)), (int(x2), int(y2)), (0, 255, 255))
        leen = len(name)*10
        leen = int(x2+leen)

        
        if acc >= 0.10:
           cv2.rectangle(draw_img, (x, y - 35), (leen, y), (0, 255, 0), cv2.FILLED)
           cv2.putText(draw_img, '{} [{:.3f}]'.format(name,acc), (x + 6, y - 6), font, 1.0, (255, 0, 0), 1)
        else:
           cv2.rectangle(draw_img, (x, y - 35), (leen-6, y), (0, 0, 255), cv2.FILLED)
           cv2.putText(draw_img, 'Unknown', (x + 6, y - 6), font, 1.0, (255, 255, 255), 1)

    cv2_imshow(draw_img)
```

```python
#Load KNN Model
with open('knn_Clasifi.pkl','rb') as f:
     knn,all_labels,unique_names = pickle.load(f)
```

```python
get_name = lambda n:unique_names[all_labels[n]] 

def classification_knn(emb):
    dis, label = knn.kneighbors(emb)
    best_class = label.ravel()[0]
    acc  = 1.0 - dis.ravel().min()
    name = get_name(best_class)
    print(f"Similar to: {name}, ACC: {acc:.2f}")
    return name, acc


#font = cv2.FONT_HERSHEY_DUPLEX
font = 5 #cv2.FONT_HERSHEY_COMPLEX

def test_one_knn(path_img):
    img = cv2.imread(path_img)
    draw_img = img.copy()
    all_boxes = face_detect(img, one_only=False)
    norf = all_boxes.shape[0]
    for box in all_boxes:
        cut_fc = load_mtcnn2.Crop_Align(img, box)
        #embed_face = FACENET.Embedding([cut_fc], MTCNN.models)
        embed_face = FACENET.Embedding(cut_fc, norf, MTCNN.models)
        name, acc = classification_knn(embed_face)
        
        x,y,x2,y2,_ = box.astype(int)
        cv2.rectangle(draw_img, (int(x), int(y)), (int(x2), int(y2)), (0, 255, 255))
        leen = len(name)*10
        leen = int(x2+leen)

        
        if acc >= 0.61:
           cv2.rectangle(draw_img, (x, y - 35), (leen, y), (0, 255, 0), cv2.FILLED)
           cv2.putText(draw_img, '{} [{:.3f}]'.format(name,acc), (x + 6, y - 6), font, 1.0, (255, 0, 0), 1)
        else:
           cv2.rectangle(draw_img, (x, y - 35), (leen-6, y), (0, 0, 255), cv2.FILLED)
           cv2.putText(draw_img, 'Unknown', (x + 6, y - 6), font, 1.0, (255, 255, 255), 1)

    cv2_imshow(draw_img)
```
